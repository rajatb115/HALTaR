## Create finetune dataset

### Read each table and perform the following augmentation steps
1. First select the random columns.
2. Select the random rows.
3. Repeat the above steps untill all the columns are covered and achieved min tables.

```python
## checking the title of the tables

for k in table_content.keys():
    for kk in table_content[k].keys():
        print(kk)
    break
```

### Augmenting the sub-tables
1. We want the augmentation such that All columns must be covered across subtables
2. Row coverage is optional / partial
3. Each subtable has a bounded number of columns and rows

```python
import random

def get_subtables(table, columns, min_tables, min_row, max_row, min_col, max_col, seed=42):

    # setting the random seed for reproducibility
    # random.seed(seed)

    total_col = len(columns)
    total_row = len(table)
    column_cover_track = [0]*total_col ## to track the coverage of the columns
    subtables = []
    
    while True:
        ## Now select the number of columns that need to be selected
        total_col_selected = random.randint(min_col, max_col)
        total_col_selected = min(total_col_selected, total_col)
        # print("total columns selected", total_col_selected)

        column_list = [i for i in range(total_col)]
        column_list_selected = random.sample(column_list, total_col_selected)

        ## Selecting the columns for the subtables
        subtables_columns = []
        for i in column_list_selected:
            subtables_columns.append(columns[i])

        ## Now select the number of rows that need to be selected
        total_row_selected = random.randint(min_row, max_row)
        total_row_selected = min(total_row_selected, total_row)

        ## Selecting the rows for the subtables
        sampled_table_rows = []
        for i in table:
            tmp_row = []
            for j in column_list_selected:
                tmp_row.append(i[j])
            sampled_table_rows.append(tmp_row)
        
        ## now selecting the rows
        row_list = [i for i in range(total_row)]
        row_list_selected = random.sample(row_list, total_row_selected)

        subtables_rows = []
        for i in row_list_selected:
            subtables_rows.append(sampled_table_rows[i])

        subtables.append([subtables_columns, subtables_rows])

        for i in column_list_selected:
            column_cover_track[i] = 1

        if sum(column_cover_track) == total_col and len(subtables) >= min_tables:
            break
    
    return subtables
```

```python
def clean_table_nan(table):
    temp_table = []

    cnt = 0
    for t in table:
        if "nan" in t:
            cnt+=1
        else:
            temp_table.append(t)
    
    print(len(table), cnt)
    return temp_table

def replace_table_nan(table):
    temp_table = []

    for t in table:
        if "nan" in t:
            tmp_list = []
            for val in t:
                if val=="nan":
                    tmp_list.append("")
                else:
                    tmp_list.append(val)
            temp_table.append(tmp_list)
        else:
            temp_table.append(t)
    return temp_table
```

```python
augmented_data = {}
total_augmented_tables = 0

for k in table_content.keys():
    print(k, end=" : ")

    table = table_content[k]["data"]
    title = table_content[k]["title"]

    # table = clean_table_nan(table)
    table = replace_table_nan(table) ## replace the nan as empty string
    
    # augmented_data[k] = get_subtables(table, title, min_tables, min_rows, max_rows, min_cols, max_cols)
    # total_augmented_tables += len(augmented_data[k])

    # print("No. of rows:", len(table), ": No. of columns:", len(title), ": Number of sub-tables:", len(augmented_data[k]))

    ## note we need atleat two tables for constucting the positive pair
    # In case there is no row we just compute the negative examples 
    if len(table) <= min_rows:
        subtable_count = min(min_tables, max(2, len(title)- min_cols))
        augmented_data[k] = get_subtables(table, title, subtable_count, min_rows, max_rows, min_cols, max_cols)
    else:
        augmented_data[k] = get_subtables(table, title, min_tables, min_rows, max_rows, min_cols, max_cols)
    
    total_augmented_tables += len(augmented_data[k])

    print("No. of rows:", len(table), ": No. of columns:", len(title), ": Number of sub-tables:", len(augmented_data[k]))
```

```python
print("Total augmented tables:", total_augmented_tables)
```

## Saving this augmented dataset for finetuning the model

```python
# table_content contain all the details about the table
# augmented_data contain all the augmented data for a table
# get the meta_data from the table_content for each table in augmented_data and change the name of the table and the table content

full_database_dic = {}

for k in augmented_data.keys():

    full_table_dic = {}

    tmp = k.strip().split('-')
    database_name = tmp[1].strip()
    table_name = tmp[2].strip()

    for idx, data in enumerate(augmented_data[k]):

        # print(data)
        if num_col_flag:
            col_list = data[0]
            num_col_list = data[1]
            table_val = data[2]
        else:
            col_list = data[0]
            table_val = data[1]
            num_col_list = []

        metadata = table_content[k].copy()
        
        # print(metadata)
        metadata['title'] = col_list
        metadata['numericColumns'] = num_col_list
        metadata['data'] = table_val
        metadata['numCols'] = len(col_list)
        metadata['numDataRows'] = len(table_val)
        
        full_table_dic[f'table-{database_name}-{table_name}_{idx+1}'] = metadata

    # print(full_table_dic)
    if f're_tables-{database_name}' not in full_database_dic.keys():
        full_database_dic[f're_tables-{database_name}'] = []
    
    full_database_dic[f're_tables-{database_name}'].append(full_table_dic)
```

## Now constructing the folds file used while finetuning the model

```python
import itertools
import random

def processed(examples, pos):
    combinations = []
    for e in examples:
        combinations.append(list(e)+[pos])
    
    return combinations

def get_positive_examples(table_names, max_exp):

    combinations = list(itertools.combinations(table_names, 2))

    if len(combinations) <= max_exp:
        return processed(combinations, '1')
    else:
        # choose max_exp from combinations
        combinations_tmp = random.sample(combinations, max_exp)
        return processed(combinations_tmp, '1')
    
def get_positive_examples_2(table_names, max_exp):
    pos_examples = []

    for exp in table_names:
        temp_table_names = list( set(table_names) - set(exp))

        pos_data = random.sample(temp_table_names, max_exp)

        for d in pos_data:
            pos_examples.append([exp, d, '1'])
    
    return pos_examples

def get_positive_examples_db(table_names, corpus, max_exp):
    pos_examples = []

    for exp in table_names:
        # selecting the max_exp number of unique elements from the corpus
        pos_data = random.sample(corpus, max_exp)

        for d in pos_data:
            pos_examples.append([exp, d, '1'])

    return pos_examples
```

```python
# construct positive examples and max positive example per table will be 10
# max_p_ex_per_table = 20
max_p_ex_per_table = 2

max_p_ex_per_db = 1

positive_examples = []

# positive examples from a single table
for k in full_database_dic.keys():

    for t in full_database_dic[k]:
        # print(len(t.keys()))
        if len(list(t.keys())) >1:
            # positive_examples += get_positive_examples(list(t.keys()), max_p_ex_per_table)
            positive_examples += get_positive_examples_2(list(t.keys()), max_p_ex_per_table)


# # positive examples from the whole database
# for k in full_database_dic.keys():
    
#     tables_of_same_db = []

#     for t in full_database_dic[k]:
#         tables_of_same_db += list(t.keys())

#     for t in full_database_dic[k]:
#         # removing the current database content
#         rem_tables_of_same_db = list( set(tables_of_same_db) - set(list(t.keys())))

#         positive_examples += get_positive_examples_db(list(t.keys()), rem_tables_of_same_db, max_p_ex_per_db)
```

```python
import random

# here I am not taking any table from the same corpus too for negative queries

def get_negative_examples(table_names, corpus, max_exp):
    corpus_tmp = list( set(corpus) - set(table_names))
    
    neg_examples = []
    
    for exp in table_names:
        neg_data = random.sample(corpus_tmp, max_exp)
        for d in neg_data:
            neg_examples.append([exp, d, '0'])

    return neg_examples
```

```python
# constructing negative examples
max_n_ex_per_table = 2

negative_examples = []

corpus_tables = []

# taking all the tables in the corpus
for k in full_database_dic.keys():
    for t in full_database_dic[k]:
        corpus_tables += list(t.keys())

for k in full_database_dic.keys():
    for t in full_database_dic[k]:
        negative_examples += get_negative_examples(list(t.keys()), corpus_tables, max_n_ex_per_table)
```

## Creating the Kfolds

```python
import numpy as np
from sklearn.model_selection import KFold  # if sklearn is allowed

n_splits = 10

# kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
kf = KFold(n_splits=n_splits, shuffle=True)
```

```python
## making sample distribution equal
data_pos = np.array(positive_examples, dtype=str)
data_neg = np.array(negative_examples, dtype=str)

data_pos_folds = kf.split(data_pos)
data_neg_folds = kf.split(data_neg)
```

```python
import random

folds = []

for data_pos_fold, data_neg_fold in zip(data_pos_folds, data_neg_folds):
    train_pos_idx, val_pos_idx = data_pos_fold
    train_neg_idx, val_neg_idx = data_neg_fold

    train_pos_set = data_pos[train_pos_idx]
    val_pos_set = data_pos[val_pos_idx]

    train_neg_set = data_neg[train_neg_idx]
    val_neg_set = data_neg[val_neg_idx]

    train_set = np.concatenate([train_pos_set, train_neg_set])
    val_set = np.concatenate([val_pos_set, val_neg_set])

    ## shuffling the train and val set

    np.random.shuffle(train_set)
    np.random.shuffle(val_set)
    
    folds.append([train_set, val_set])
```

```python

# save
np.save(f"bird_similarity_folds_{run_number}_mint_{min_tables}_minr_{min_rows}_maxr_{max_rows}_minc_{min_cols}_maxc_{max_cols}.npy", folds)
```
